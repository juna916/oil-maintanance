```python
import heapq

class OilFieldManager:
    def __init__(self):
        self.oil_wells = []
        self.maintenance_queue = []
        self.well_id_counter = 1

    def add_oil_well(self, name, location, status, daily_production):
        oil_well = {
            "id": self.well_id_counter,
            "name": name,
            "location": location,
            "status": status,
            "daily_production": daily_production
        }
        self.oil_wells.append(oil_well)
        self.well_id_counter += 1
        print(f"Sumur minyak '{name}' berhasil ditambahkan.")

    def remove_oil_well(self, well_id):
        for well in self.oil_wells:
            if well["id"] == well_id:
                self.oil_wells.remove(well)
                print(f"Sumur minyak '{well['name']}' berhasil dihapus.")
                return
        print("ID sumur minyak tidak ditemukan.")

    def schedule_maintenance(self, well_id, priority):
        for well in self.oil_wells:
            if well["id"] == well_id:
                heapq.heappush(self.maintenance_queue, (priority, well_id, well["name"]))
                print(f"Perawatan untuk sumur '{well['name']}' berhasil dijadwalkan dengan prioritas {priority}.")
                return
        print("ID sumur minyak tidak ditemukan.")

    def process_maintenance(self):
        if self.maintenance_queue:
            priority, well_id, name = heapq.heappop(self.maintenance_queue)
            print(f"Memproses perawatan untuk sumur '{name}' (Prioritas: {priority}).")
        else:
            print("Tidak ada perawatan yang dijadwalkan.")

    def display_production_report(self):
        total_production = sum(well["daily_production"] for well in self.oil_wells if well["status"] == "aktif")
        print(f"Total produksi harian dari semua sumur aktif: {total_production} barel.")

    def display_oil_wells(self):
        if self.oil_wells:
            print("\nDaftar Sumur Minyak:")
            for well in self.oil_wells:
                print(f"ID: {well['id']}, Nama: {well['name']}, Lokasi: {well['location']}, Status: {well['status']}, Produksi Harian: {well['daily_production']} barel")
        else:
            print("Belum ada data sumur minyak.")

# Main Program
manager = OilFieldManager()

while True:
    print("\nMenu:")
    print("1. Tambah Sumur Minyak")
    print("2. Hapus Sumur Minyak")
    print("3. Jadwalkan Perawatan")
    print("4. Proses Perawatan")
    print("5. Tampilkan Laporan Produksi")
    print("6. Tampilkan Daftar Sumur")
    print("7. Keluar")

    choice = input("Pilih menu: ")

    if choice == "1":
        name = input("Nama Sumur: ")
        location = input("Lokasi Sumur: ")
        status = input("Status (aktif/non-aktif): ")
        daily_production = int(input("Produksi Harian (barel): "))
        manager.add_oil_well(name, location, status, daily_production)
    elif choice == "2":
        well_id = int(input("Masukkan ID sumur yang ingin dihapus: "))
        manager.remove_oil_well(well_id)
    elif choice == "3":
        well_id = int(input("Masukkan ID sumur: "))
        priority = int(input("Masukkan prioritas (1=tinggi, 2=sedang, 3=rendah): "))
        manager.schedule_maintenance(well_id, priority)
    elif choice == "4":
        manager.process_maintenance()
    elif choice == "5":
        manager.display_production_report()
    elif choice == "6":
        manager.display_oil_wells()
    elif choice == "7":
        print("Keluar dari sistem. Terima kasih!")
        break
    else:
        print("Pilihan tidak valid, coba lagi.")
```
